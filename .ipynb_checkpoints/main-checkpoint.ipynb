```python
import pandas as pd
import numpy as np
```

```python
train1 = pd.read_excel('data/train.xlsx').iloc[:, :2]
train1.columns = ["question", "answer"]
train2 = pd.read_excel('data/test.xlsx').iloc[:, :2]
train2.columns = ["question", "answer"]
train = pd.concat([train1, train2],axis=0, ignore_index=True)
train
```

```python
train.drop_duplicates(subset="question", keep='first',inplace=True)
# 添加一些数据处理
train
```

```python
train.to_json('data/processedData.txt')
```
