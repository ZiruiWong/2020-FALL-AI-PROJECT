```python
import pandas as pd
train1 = pd.read_excel('data/data1.xlsx', engine='openpyxl').iloc[:, :2]
train1.columns = ["question", "answer"]

# 取data2的问答对
train2 = pd.read_excel('data/data2.xlsx', engine='openpyxl').iloc[:, :2]
train2.columns = ["question", "answer"]

# 数据处理
        # 将数据进行拼接 并去除问题重复数据
train = pd.concat([train1, train2], axis=0, ignore_index=True)  # 拼接数据
train.drop_duplicates(subset="question", keep='first', inplace=True)  # 数据去重
qTrain = train["question"]
aTrain = train["answer"]
qTrain
```

```python
import pkuseg
import sys
import unicodedata as ud
seg = pkuseg.pkuseg()
stopwords = [line.strip() for line in open('./dic/stopword.txt', 'r').readlines()]
punctuation = dict.fromkeys(i for i in range(sys.maxunicode) if ud.category(chr(i)).startswith('P'))

def cut1Sentence(sentence):
    print(sentence)
    sentence = sentence.strip()  # 去除字符串两端空格
    sentence = sentence.translate(punctuation)  # 去除标点符号
    
    seg_list = seg.cut(sentence)
    results = []
    for word in seg_list:
        if word not in stopwords:
            results.append(word)
    return results

cutSentences = [cut1Sentence(sentence) for sentence in qTrain[:100]]
cutSentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
document = [" ".join(sent0) for sent0 in cutSentences]
tfidf_model = TfidfVectorizer(token_pattern=r"(?u)\b\w+\b").fit_transform(document)
tfidf_model
```

```python
import pkuseg
import sys
import unicodedata as ud
punctuation = dict.fromkeys(i for i in range(sys.maxunicode) if ud.category(chr(i)).startswith('P'))

wordsList = []
seg = pkuseg.pkuseg()           # 以默认配置加载模型
for sentence in qTrain:
    sentence = str(sentence)
    sentence = sentence.translate(punctuation)  # 去除标点符号
    seg_list = seg.cut(sentence)
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word)
    wordsList.append(results)
wordsList[:10]
```

```python
from collections import defaultdict
frequency = defaultdict(int)
for text in wordsList:
     for token in text:
        frequency[token] += 1
frequency
```

```python
wordsList = [[token for token in text if frequency[token] > 1] for text in wordsList]
wordsList[:10]
```

```python
from gensim import corpora, models, similarities
dictionary = corpora.Dictionary(wordsList)
dictionary
```

```python
corpus_simple = [dictionary.doc2bow(text) for text in wordsList]
corpus_simple
```

```python
model = models.TfidfModel(corpus_simple)
model.save("model/word2vec.model")
corpus = model[corpus_simple]
index = similarities.Similarity(model, corpus, len(dictionary))
index
```
