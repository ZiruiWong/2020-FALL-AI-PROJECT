```python
import pandas as pd
train1 = pd.read_excel('data/data1.xlsx', engine='openpyxl').iloc[:, :2]
train1.columns = ["question", "answer"]

# 取data2的问答对
train2 = pd.read_excel('data/data2.xlsx', engine='openpyxl').iloc[:, :2]
train2.columns = ["question", "answer"]

# 数据处理
# 将数据进行拼接 并去除问题重复数据
train = pd.concat([train1, train2], axis=0, ignore_index=True)  # 拼接数据
train.drop_duplicates(subset="question", keep='first', inplace=True)  # 数据去重
train.reset_index(inplace=True)

qTrain = train["question"]
aTrain = train["answer"]
print(qTrain)
```

```python
stopwords = [line.strip() for line in open('./dic/stopword.txt', 'r',encoding = 'utf-8').readlines()]
stopwords
```

```python
import jieba

for sent in qTrain[:100]:
    seg_list = jieba.lcut(sent)
    results = []
    for seg in seg_list:            
        if seg not in stopwords:    
            results.append(seg)
    print(results)
```

```python
# pkuseg
import pkuseg
import sys
import unicodedata as ud
punctuation = dict.fromkeys(i for i in range(sys.maxunicode) if ud.category(chr(i)).startswith('P'))

seg = pkuseg.pkuseg()           # 以默认配置加载模型
sentence = "第一届冬奥会在哪举办？"
sentence = sentence.translate(punctuation)  # 去除标点符号
seg_list = seg.cut(sentence)
results = []
for word in seg_list:               
    if word not in stopwords:
            results.append(word)
print(results)
```

```python
# pkuseg
import pkuseg
import sys
import unicodedata as ud
punctuation = dict.fromkeys(i for i in range(sys.maxunicode) if ud.category(chr(i)).startswith('P'))

seg = pkuseg.pkuseg()           # 以默认配置加载模型
for sentence in qTrain[:100]:
    seg_list = seg.cut(sentence)
    results = []
    for word in seg_list:        
        word = word.translate(punctuation)  # 去除标点符号
        if word not in stopwords :   
            results.append(word)
    print(results)
```

```python
# pkuseg
import pkuseg

seg = pkuseg.pkuseg(user_dict = './dic/dictionary.txt')           # 以默认配置加载模型
for sentence in qTrain[:100]:
    seg_list = seg.cut(sentence)
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word)
    print(results)
```

```python
# jiagu
import jiagu

jiagu.init()
for sentence in qTrain[:100]:
    seg_list = jiagu.seg(sentence)
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word)
    print(results)
```

```python
# 
import thulac

thu = thulac.thulac(seg_only=True)
for sentence in qTrain[:100]:
    seg_list = thu.cut(sentence, text=True).split()
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word)
    print(results)
```

```python
import pynlpir

pynlpir.open()

for sentence in qTrain[:100]:
    seg_list = pynlpir.segment(sentence, pos_tagging=False)
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word)
    print(results)

pynlpir.close()
```

```python
import pyhanlp

for sentence in qTrain[:100]:
    seg_list = pyhanlp.HanLP.segment(sentence)
    results = []
    for word in seg_list:            
        if word not in stopwords:    
            results.append(word.word)
    print(results)
```
