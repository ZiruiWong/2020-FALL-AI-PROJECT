```python
from smoothnlp.algorithm import kg
from kgexplore import visual
ngrams = kg.extract_ngram(["SmoothNLPV0.3",
                            "SmoothNLPNLP",
                            "SmoothNLPPythonJava",
                            "SmoothNLP",
                            "SmoothNLP",
                            "SmoothNLPV0.4",
                            "KGExploreSmoothNLP"])
visual.visualize(ngrams,width=12,height=10)
```
